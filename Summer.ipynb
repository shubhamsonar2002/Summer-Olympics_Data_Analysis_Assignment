# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd

df = pd.read_csv("summer.csv")
a = df['City'].drop_duplicates()
no_of_cities = len(a)

print(f"Number of cities summer olympics is held are {no_of_cities} cities")
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
import matplotlib.pyplot as plt


medals = df['Medal']
sports = df['Sport']
unique_sport = sports.drop_duplicates()
dict_sport = {}


for i in unique_sport:
    dict_sport[str(i)] = 0
    
for i in range(len(medals)):
    if medals[i]=='Gold':
        dict_sport[sports[i]] += 1

max_5keys_2 = sorted(dict_sport, key=dict_sport.get, reverse=True)[:5]
max_5values_2 = [dict_sport[i] for i in max_5keys_2]

plt.title("Sport having most number of gold medals")
plt.barh(max_5keys_2,max_5values_2)
plt.show()
```
